```
# Mark Green Script !python 3
# Scrapes the USNews site for best Nursing Graduate school rankings

import requests 
import re
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep, time
from random import randint
from warnings import warn


# Declare the lists to store the data
colleges = []
locations = []
rank = []
cost = []
enrollment = []
coltype = []

# Prepare the monitoring of the loop
start_time = time()
rqs = 0

# headers so we can access the site
headers = {
    'User-Agent': 'My User Agent 1.0',
    'From': 'youremail@domain.com' 
}

pages_url = [str(i) for i in range(1,10)]

#url= 'https://www.usnews.com/education/online-education/nursing/rankings'
#for every page in the list
for page_url in pages_url:
    response = requests.get('https://www.usnews.com/education/online-education/nursing/rankings?_page=' + str(page_url), headers=headers)
    sleep(randint(8,15))

    rqs += 1
    elapsed_time = time() - start_time
    print('Request: {}; Frequency: {} requests/s'.format(rqs, rqs/elapsed_time))


    if response.status_code != 200:
        warn('Request: {}; Status code: {}'.format(rqs, response.status_code))

        # Break the loop if the number of requests is greater than expected
        #if requests > 200:
        warn('Number of requests was greater than expected.')  
        break 

        # Parse the content of the request with BeautifulSoup
    html_soup = BeautifulSoup(response.text, 'html.parser')


    coll_containers = html_soup.find_all('li', {'data-view':'online-education-search-ranking-results-card'})

    for container in coll_containers:
        try:
            name = container.h3.a.text
            colleges.append(name)

            loc = container.find('div', class_='block-normal text-small').text
            locations.append(loc)

            rk = re.sub(r'[\ \n]{2,}', '',container.find('div', class_='text-strong').text)
            rank.append(rk)
        
            ct = container.find('div', class_= 'display-inline-for-medium-up inline-right-tight-for-medium-up border-right-for-medium-up').strong
            ct = ct.text
            cost.append(ct)
        
            enroll = container.find('div', class_= 'display-inline-for-medium-up inline-right-tight-for-medium-up border-right-for-medium-up').next_sibling.next_sibling.strong
            enroll = enroll.text
            enrollment.append(enroll)

            tp = container.find('div', class_= 'display-inline-for-medium-up inline-right-tight-for-medium-up border-right-for-medium-up').next_sibling.next_sibling.next_sibling.next_sibling.strong
            tp = tp.text
            coltype.append(tp)

        except AttributeError:
            colleges.append(None)

onlinenurs_df = pd.DataFrame({'college':colleges,
                        'location': locations,
                        'rank' : rank,
                        'cost': cost,
                        'enrollment': enrollment})
print(onlinenurs_df.info())
print(onlinenurs_df)
#print(len(coll_containers))

 
    
```
